```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# step astivation function
def step_function(z):
  return 1 if z >= 0 else 0
```

```python
# perceptron algorithm
def perceptron(x, y, learning_rate=0.1, epochs=10):
  weights = np.zeros(x.shape[1])
  bias = 0
  for epoch in range(epochs):
    for i in range(len(x)):
      z = np.dot(x[i], weights)+bias
      y_pred = step_function(z)
      error = y[i]-y_pred
      if error!=0:
        weights+=learning_rate*error*x[i]
        bias+=learning_rate*error
    return weights, bias
```

```python
# Dataset
X = np.array([
    [2.0, 1.0],
    [1.0, 1.0],
    [2.0, 3.0],
    [3.0, 2.0],
    [1.5, 0.5],
    [1.0, 2.0],
    [3.0, 1.5],
    [2.5, 2.0]
])
```

```python
y = np.array([1, 0, 1, 1, 0, 0, 1, 1])  # Labels
```

```python
# Train the perceptron
learning_rate = 0.1
epochs = 10
weights, bias = perceptron(X, y, learning_rate, epochs)
```

```python
# Plotting the dataset and decision boundary
def plot_decision_boundary(X, y, weights, bias):
    # Plot the points
    for i in range(len(X)):
        if y[i] == 1:
            plt.scatter(X[i][0], X[i][1], color='blue', label='Class 1' if i == 0 else "")
        else:
            plt.scatter(X[i][0], X[i][1], color='red', label='Class 0' if i == 1 else "")
```

```python
# prompt:  # Plot the decision boundary
#     x_min, x_max = min(X[:, 0]) - 1, max(X[:, 0]) + 1
#     y_min, y_max = min(X[:, 1]) - 1, max(X[:, 1]) + 1
#     x_values = np.linspace(x_min, x_max, 100)
#     y_values = -(weights[0] * x_values + bias) / weights[1]  # Equation of the decision boundary
#     plt.plot(x_values, y_values, color='green', label='Decision Boundary')
#     plt.xlim(x_min, x_max)
#     plt.ylim(y_min, y_max)
#     plt.xlabel('x1')
#     plt.ylabel('x2')
#     plt.legend()
#     plt.title('Perceptron Algorithm - Decision Boundary')
#     plt.grid(True)
#     plt.show()
# plot_decision_boundary(X, y, weights, bias)

import numpy as np
import matplotlib.pyplot as plt

# step astivation function
def step_function(z):
  return 1 if z >= 0 else 0

# perceptron algorithm
def perceptron(x, y, learning_rate=0.1, epochs=10):
  weights = np.zeros(x.shape[1])
  bias = 0
  for epoch in range(epochs):
    for i in range(len(x)):
      z = np.dot(x[i], weights)+bias
      y_pred = step_function(z)
      error = y[i]-y_pred
      if error!=0:
        weights+=learning_rate*error*x[i]
        bias+=learning_rate*error
  return weights, bias

# Dataset
X = np.array([
    [2.0, 1.0],
    [1.0, 1.0],
    [2.0, 3.0],
    [3.0, 2.0],
    [1.5, 0.5],
    [1.0, 2.0],
    [3.0, 1.5],
    [2.5, 2.0]
])
y = np.array([1, 0, 1, 1, 0, 0, 1, 1])  # Labels

# Train the perceptron
learning_rate = 0.1
epochs = 10
weights, bias = perceptron(X, y, learning_rate, epochs)

# Plotting the dataset and decision boundary
def plot_decision_boundary(X, y, weights, bias):
    # Plot the points
    for i in range(len(X)):
        if y[i] == 1:
            plt.scatter(X[i][0], X[i][1], color='blue', label='Class 1' if i == 0 else "")
        else:
            plt.scatter(X[i][0], X[i][1], color='red', label='Class 0' if i == 1 else "")

    # Plot the decision boundary
    x_min, x_max = min(X[:, 0]) - 1, max(X[:, 0]) + 1
    y_min, y_max = min(X[:, 1]) - 1, max(X[:, 1]) + 1
    x_values = np.linspace(x_min, x_max, 100)
    y_values = -(weights[0] * x_values + bias) / weights[1]  # Equation of the decision boundary
    plt.plot(x_values, y_values, color='green', label='Decision Boundary')
    plt.xlim(x_min, x_max)
    plt.ylim(y_min, y_max)
    plt.xlabel('x1')
    plt.ylabel('x2')
    plt.legend()
    plt.title('Perceptron Algorithm - Decision Boundary')
    plt.grid(True)
    plt.show()

plot_decision_boundary(X, y, weights, bias)
```
