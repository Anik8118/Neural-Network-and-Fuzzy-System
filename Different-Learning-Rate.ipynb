```python
import numpy as np
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])
```

```python

theta0 = 0
theta1 = 0
iterations = 150
m = len(x)
```

```python

def compute_cost(theta0, theta1, x, y):
    m = len(y)
    predictions = theta0 + theta1 * x
    cost = (1 / (2 * m)) * np.sum((predictions - y) ** 2)
    return cost
```

```python
learning_rates = [0.01, 0.05, 0.1]
```

```python
def gradient_descent(learning_rate):
    theta0 = 0
    theta1 = 0
    cost_history = []

    for i in range(iterations):
        predictions = theta0 + theta1 * x


        d_theta0 = (1 / m) * np.sum(predictions - y)
        d_theta1 = (1 / m) * np.sum((predictions - y) * x)

        theta0 = theta0 - learning_rate * d_theta0
        theta1 = theta1 - learning_rate * d_theta1

        cost = compute_cost(theta0, theta1, x, y)
        cost_history.append(cost)

    return cost_history
```

```python
import matplotlib.pyplot as plt
for lr in learning_rates:
    cost_history = gradient_descent(lr)
    plt.plot(range(iterations), cost_history, label=f'Learning Rate = {lr}')

plt.xlabel("Iterations")
plt.ylabel("Cost Function J(Î¸)")
plt.title("Effect of Learning Rate on Gradient Descent Optimization")
plt.legend()
plt.grid(True)
plt.show()
```
