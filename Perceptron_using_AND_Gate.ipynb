```python
# Define the AND gate dataset
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 0, 0, 1])
```

```python
# Step activation function
def step_function(x):
    return np.where(x >= 0, 1, 0)
```

```python
# Sigmoid activation function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
# Perceptron algorithm
def perceptron(X, y, activation_func, epochs=100, learning_rate=0.1):
    weights = np.zeros(X.shape[1])
    bias = 0
    for _ in range(epochs):
        for i in range(X.shape[0]):
            linear_output = np.dot(X[i], weights) + bias
            prediction = activation_func(linear_output)
            error = y[i] - prediction
            weights += learning_rate * error * X[i]
            bias += learning_rate * error
    return weights, bias
```

```python
# Train perceptron with step activation
weights_step, bias_step = perceptron(X, y, step_function)
```

```python
# Train perceptron with sigmoid activation
weights_sigmoid, bias_sigmoid = perceptron(X, y, lambda x: sigmoid(x) > 0.5)
```

```python
# Plot decision boundaries
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
x1_range = np.linspace(-0.5, 1.5, 100)
x2_range = np.linspace(-0.5, 1.5, 100)
xx, yy = np.meshgrid(x1_range, x2_range)

def plot_decision_boundary(ax, weights, bias, activation_func, title):
    Z = np.zeros(xx.shape)
    for i in range(xx.shape[0]):
        for j in range(xx.shape[1]):
            point = np.array([xx[i, j], yy[i, j]])
            linear_output = np.dot(point, weights) + bias
            Z[i, j] = activation_func(linear_output)
    ax.contourf(xx, yy, Z, levels=[-0.1, 0.5, 1.1], alpha=0.8, cmap="coolwarm")
    ax.scatter(X[:, 0], X[:, 1], c=y, cmap="coolwarm", edgecolors="k")
    ax.set_title(title)
    ax.set_xlabel("x1")
    ax.set_ylabel("x2")
    ax.set_xlim(-0.5, 1.5)
    ax.set_ylim(-0.5, 1.5)
```

```python
# Plot for step function
plot_decision_boundary(axes[0], weights_step, bias_step, step_function, "Step Function Activation")
```

```python
# Plot for sigmoid function
plot_decision_boundary(axes[1], weights_sigmoid, bias_sigmoid, sigmoid, "Sigmoid Function Activation")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the AND gate dataset
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 0, 0, 1])

# Step activation function
def step_function(x):
    return np.where(x >= 0, 1, 0)

# Sigmoid activation function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Perceptron algorithm
def perceptron(X, y, activation_func, epochs=100, learning_rate=0.1):
    weights = np.zeros(X.shape[1])
    bias = 0
    for _ in range(epochs):
        for i in range(X.shape[0]):
            linear_output = np.dot(X[i], weights) + bias
            prediction = activation_func(linear_output)
            error = y[i] - prediction
            weights += learning_rate * error * X[i]
            bias += learning_rate * error
    return weights, bias

# Train perceptron with step activation
weights_step, bias_step = perceptron(X, y, step_function)

# Train perceptron with sigmoid activation
weights_sigmoid, bias_sigmoid = perceptron(X, y, lambda x: sigmoid(x) > 0.5)

# Plot decision boundaries
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
x1_range = np.linspace(-0.5, 1.5, 100)
x2_range = np.linspace(-0.5, 1.5, 100)
xx, yy = np.meshgrid(x1_range, x2_range)

def plot_decision_boundary(ax, weights, bias, activation_func, title):
    Z = np.zeros(xx.shape)
    for i in range(xx.shape[0]):
        for j in range(xx.shape[1]):
            point = np.array([xx[i, j], yy[i, j]])
            linear_output = np.dot(point, weights) + bias
            Z[i, j] = activation_func(linear_output)
    ax.contourf(xx, yy, Z, levels=[-0.1, 0.5, 1.1], alpha=0.8, cmap="coolwarm")
    ax.scatter(X[:, 0], X[:, 1], c=y, cmap="coolwarm", edgecolors="k")
    ax.set_title(title)
    ax.set_xlabel("x1")
    ax.set_ylabel("x2")
    ax.set_xlim(-0.5, 1.5)
    ax.set_ylim(-0.5, 1.5)

# Plot for step function
plot_decision_boundary(axes[0], weights_step, bias_step, step_function, "Step Function Activation")

# Plot for sigmoid function
plot_decision_boundary(axes[1], weights_sigmoid, bias_sigmoid, sigmoid, "Sigmoid Function Activation")

plt.tight_layout()
plt.show()
```
